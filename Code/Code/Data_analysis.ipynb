# Createing a function for easy plot of categorical variables

```python
def cat_variable_plot(p_var_name, p_data):
    nr_obs = p_data.shape[0]
    var_count = p_data[p_var_name].value_counts()
    print(var_count)
    #first barplot
    (var_count/nr_obs).plot.bar(rot = 0, figsize = (15,7.5))
    
    #second barplot
    ret_0 = p_data.loc[p_data['Retained'] == 0, p_var_name ].value_counts()/var_count
    ret_1 = p_data.loc[p_data['Retained'] == 1, p_var_name ].value_counts()/var_count

    df = pd.DataFrame({'Retained_0': ret_0,
                   'Retained_1': ret_1,
                   'Count' : var_count})
    ax = df.sort_values(by='Count',ascending=False)

    ax[['Retained_1', 'Retained_0']].plot.bar(rot=0, stacked = True, figsize = (15,7.5))
```
